```python
# Importing Libraries

import pandas as pd
import unicodedata
```

```python
# Reading files

customers=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_customers_dataset.csv")
geolocation=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_geolocation_dataset.csv")
order_items=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_order_items_dataset.csv")
order_payments=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_order_payments_dataset.csv")
order_reviews=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_order_reviews_dataset.csv")
orders=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_orders_dataset.csv")
products=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_products_dataset.csv")
sellers=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\olist_sellers_dataset.csv")
product_category_name_translation=pd.read_csv("C:\\Users\\admi\\Desktop\\Saachi\\Projects\\E-commerce Customer and Sales Analytics Platform\\data\\raw\\product_category_name_translation.csv")
```

```python
# Converting timestamps

order_items["shipping_limit_date"] = pd.to_datetime(order_items["shipping_limit_date"])
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_approved_at"] = pd.to_datetime(orders["order_approved_at"])
orders["order_delivered_carrier_date"] = pd.to_datetime(orders["order_delivered_carrier_date"])
orders["order_delivered_customer_date"] = pd.to_datetime(orders["order_delivered_customer_date"])
orders["order_estimated_delivery_date"] = pd.to_datetime(orders["order_estimated_delivery_date"])
order_reviews["review_creation_date"] = pd.to_datetime(order_reviews["review_creation_date"])
order_reviews["review_answer_timestamp"] = pd.to_datetime(order_reviews["review_answer_timestamp"])
```

```python
# Normalizing cities

def normalize_city(city):
    city = str(city).lower().strip()

    city = unicodedata.normalize('NFKD', city)
    city = ''.join(c for c in city if not unicodedata.combining(c))

    return city

geolocation['geolocation_city'] = (geolocation['geolocation_city'].apply(normalize_city))
```

```python
# Fixing product category names

product_category_name_translation['product_category_name_english'] = (
    product_category_name_translation['product_category_name_english']
    .replace({
        'costruction_tools_garden': 'construction_tools_garden',
        'costruction_tools_tools': 'construction_tools_tools',
        'fashio_female_clothing': 'fashion_female_clothing',
        'home_confort': 'home_comfort',
        'home_comfort_2': 'home_comfort',
        'market_place': 'marketplace',
        'fixed_telephony': 'fixed_telephony_services',
        'security_and_services': 'security_services',
        'cine_photo': 'cinema_photo',
        'cds_dvds_musicals': 'cds_dvds_musicals_media',
        'la_cuisine': 'kitchen_premium',
        'food_drink': 'food_and_drink',
        'home_appliances_2': 'home_appliances'
    })
)
```

```python
# Translating product category name to english

products = products.merge(product_category_name_translation,on="product_category_name",how="left")
```

```python
# Checking product category names not translated

products[products['product_category_name_english'].isna()]['product_category_name'].unique()
```

```python
# Translating the remaining product category names

products['product_category_name_english'] = (
    products['product_category_name_english']
    .fillna(products['product_category_name'])
)

products['product_category_name_english'] = (
    products['product_category_name_english']
    .replace({
        'portateis_cozinha_e_preparadores_de_alimentos': 'kitchen_appliances_and_food_prep',
        'pc_gamer': 'gaming_pc'
    })
)
```

```python
# Changing column name

products['product_category_name'] = (products['product_category_name_english'])
products.drop(columns=['product_category_name_english'],inplace=True)
```

```python
# Correcting column names

products.rename(columns={
    'product_name_lenght': 'product_name_length',
    'product_description_lenght': 'product_description_length'
}, inplace=True)
```

```python
# Saving cleaned data

product_category_name_translation.to_csv("product_category_name_translation_clean.csv",index=False)

products.to_csv("olist_products_dataset_clean.csv",index=False)

geolocation.to_csv("olist_geolocation_dataset_clean.csv",index=False)

order_items.to_csv("olist_order_items_dataset_clean.csv",index=False)

print("All cleaned files saved successfully!")
```
